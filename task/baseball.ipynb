타자 선형회귀 모델

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Splitting batter data into train and test sets
X_batter = batter_data.drop('Salary', axis=1)
y_batter = batter_data['Salary']
X_batter_train, X_batter_test, y_batter_train, y_batter_test = train_test_split(X_batter, y_batter, test_size=0.2, random_state=42)

# Identifying numerical and categorical columns
numerical_cols_batter = X_batter.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols_batter = X_batter.select_dtypes(include=['object']).columns.tolist()

# Creating a column transformer for preprocessing
preprocessor_batter = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols_batter),
        ('cat', OneHotEncoder(drop='first'), categorical_cols_batter)
    ])

# Creating a pipeline with preprocessing and the model
pipeline_batter = Pipeline([
    ('preprocessor', preprocessor_batter),
    ('model', LinearRegression())
])

# Training the model
pipeline_batter.fit(X_batter_train, y_batter_train)

# Get R^2 score for training set
r2_score_batter_train = pipeline_batter.score(X_batter_train, y_batter_train)

r2_score_batter_train
```

##### 타자 선형회귀 시각화

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_predictions(y_true, y_pred, title):
    """Plot actual vs predicted values and residuals distribution."""
    residuals = y_true - y_pred

    fig, ax = plt.subplots(1, 2, figsize=(15, 6))
    
    # Actual vs Predicted scatter plot
    ax[0].scatter(y_true, y_pred, alpha=0.5)
    ax[0].plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], '--', lw=2, color='red')
    ax[0].set_xlabel('Actual Values')
    ax[0].set_ylabel('Predicted Values')
    ax[0].set_title(f'Actual vs Predicted - {title}')

    # Residuals distribution plot
    ax[1].hist(residuals, bins=30, alpha=0.75)
    ax[1].set_xlabel('Residuals')
    ax[1].set_ylabel('Frequency')
    ax[1].set_title(f'Residuals Distribution - {title}')

    plt.tight_layout()
    plt.show()

# Predictions for batter data using Linear Regression
y_batter_pred_linear = pipeline_batter.predict(X_batter_test)

# Plotting
plot_predictions(y_batter_test, y_batter_pred_linear, "Batter Linear Regression")
```
