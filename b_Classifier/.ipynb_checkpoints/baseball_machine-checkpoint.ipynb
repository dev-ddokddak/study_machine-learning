```python
columns_to_drop = ['WAR','WAR2', 'wOBA', 'wRC+', '타석', '타수', '연도', '순', '이름']

# 열 삭제
player_df = player_df.drop(columns=columns_to_drop)
```

```python
player_df
```

```python
player_df = player_df.rename(columns={"-": "G"})
```

```python
player_df = player_df.drop(columns='--')
```

```python
column_mapping = {
    "순": "Rank",
    "이름": "Name",
    "G": "Games",
    "득점": "Runs",
    "안타": "Hits",
    "2루타": "Doubles",
    "3루타": "Triples",
    "홈런": "HomeRuns",
    "루타": "TotalBases",
    "타점": "RBIs",
    "도루": "StolenBases",
    "도루실패": "CaughtStealing",
    "볼넷": "Walks",
    "사구": "HitByPitch",
    "고의사구": "IntentionalWalks",
    "삼진": "Strikeouts",
    "병살": "GroundIntoDoublePlay",
    "희생타": "SacrificeHits",
    "희생플라이": "SacrificeFlies",
    "타율": "BattingAverage",
    "출루": "OnBasePercentage",
    "장타": "SluggingPercentage",
    "OPS": "OPS"
}

player_df.rename(columns=column_mapping, inplace=True)

print(player_df.columns)
```

```python
player_df.info()
```

```python
import numpy as np

# "타율"과 "출루"에서 공백 값을 NaN으로 변경
player_df["BattingAverage"] = player_df["BattingAverage"].replace('  ', np.nan)
player_df["OnBasePercentage"] = player_df["OnBasePercentage"].replace('  ', np.nan)
player_df["SluggingPercentage"] = player_df["SluggingPercentage"].replace('  ', np.nan)


# NaN 값을 다루기 위한 방법 선택: NaN을 평균 값으로 대체하거나 행 삭제
# 이 예에서는 평균 값으로 대체하는 방법을 선택
player_df["BattingAverage"].fillna(player_df["BattingAverage"].astype(float).mean(), inplace=True)
player_df["OnBasePercentage"].fillna(player_df["OnBasePercentage"].astype(float).mean(), inplace=True)
player_df["SluggingPercentage"].fillna(player_df["SluggingPercentage"].astype(float).mean(), inplace=True)


# 이제 변환 수행
player_df["BattingAverage"] = player_df["BattingAverage"].astype(float).astype(int)
player_df["OnBasePercentage"] = player_df["OnBasePercentage"].astype(float).astype(int)
player_df["SluggingPercentage"] = player_df["SluggingPercentage"].astype(float).astype(int)
```

```python
# 이후 원래 코드를 다시 실행
cols_to_normalize = [col for col in player_df.columns if col not in ["Name", "Games", "OPS"]]
for col in cols_to_normalize:
    player_df[col] = player_df[col] / player_df["Games"]

player_df  # 결과 확인을 위한 출력
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVR  # SVR로 변경
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import Pipeline

# C: 하드 마진, 소프트 마진
# gamma: 하나의 훈련 샘플이 미치는 영향의 범위
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf']
}

# 서포트 벡터 회귀기
support_vector_regressor = SVR()

# 하이퍼 파라미터 튜닝
grid_support_vector = GridSearchCV(support_vector_regressor, param_grid=param_grid, cv=3, refit=True, return_train_score=True)

features, targets = player_df.iloc[:, :-1], player_df.TotalBases

X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.3)

sv_pipeline = Pipeline([('normalizer', Normalizer()), ('support_vector_regressor', grid_support_vector)])
sv_pipeline.fit(X_train, y_train)
```

```python
# DataFrame으로 변환
scores_df = pd.DataFrame(grid_support_vector.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', 
           'split0_test_score', 'split1_test_score', 'split2_test_score']].sort_values(by='rank_test_score')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def get_evaluation(y_true, y_pred):
    # 평가 지표 계산
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False) # RMSE는 squared=False로 설정
    r2 = r2_score(y_true, y_pred)
    
    # 평가 지표 출력
    # MAE (Mean Absolute Error):
    # 정의: 예측값과 실제 값의 절대 차이의 평균
    # MSE (Mean Squared Error):
    # 정의: 예측값과 실제 값의 차이를 제곱하여 평균한 값
    # RMSE (Root Mean Squared Error):
    # 정의: MSE의 제곱근 값. MSE에 비해 큰 오차에 더 큰 페널티를 부여
    # R-squared (결정 계수):
    # 정의: 회귀 모델의 설명력을 나타내는 지표. 전체 변동 중 모델이 설명하는 변동의 비율을 나타냄
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R-squared:', r2)

```

```python
prediction = sv_pipeline.predict(X_test)
get_evaluation(y_test, prediction)
```
