##### GridSearchCV
- êµì°¨ ê²€ì¦ê³¼ ìµœì ì˜ í•˜ì´í¼ íŒŒë¼ë¯¸í„° íŠœë‹ì„ í•œ ë²ˆì— í•  ìˆ˜ ìˆëŠ” ê°ì²´ì´ë‹¤.
- max_depthì™€ min_samples_splitì— 1ì°¨ì› ì •ìˆ˜í˜• listë¥¼ ì „ë‹¬í•˜ë©´, 2ì°¨ì›ìœ¼ë¡œ ê²°í•©í•˜ì—¬ ê²©ì(Grid)ë¥¼ ë§Œë“¤ê³ , ì´ ì¤‘ ìµœì ì˜ ì ì„ ì°¾ì•„ë‚¸ë‹¤.
- ë”¥ëŸ¬ë‹ì—ì„œëŠ” í•™ìŠµ ì†ë„ê°€ ë¨¸ì‹ ëŸ¬ë‹ì— ë¹„í•´ ëŠë¦¬ê³ , ë ˆì´ì–´(ì¸µ)ê°€ ê¹Šì–´ì§ˆ ìˆ˜ë¡ ì¡°ì •í•´ì£¼ì–´ì•¼ í•  í•˜ì´í¼ íŒŒë¼ë¯¸í„° ê°’ì´ ë§ì•„ì§€ê¸° ë•Œë¬¸ì—, RandomSearchCVì—ì„œ ëŒ€ëµì ì¸ ë²”ìœ„ë¥¼ ì°¾ì€ ë‹¤ìŒ, GridSearchCVë¡œ ë””í…Œì¼ì„ ì¡°ì •í•˜ëŠ” ë°©ì‹ì„ ì‚¬ìš©í•œë‹¤.

##### ë¶“ê½ƒ ë°ì´í„°ë¡œ êµì°¨ ê²€ì¦

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
features, targets = iris.data, iris.target

target_df = pd.DataFrame(targets, columns=['target'])
target_df.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

decision_tree_classifier = DecisionTreeClassifier(min_samples_leaf=6, random_state=124)
kfold = KFold(n_splits=5)
```

##### KFold.split(feature)
- featuresë§Œ ì „ë‹¬í•˜ê³  í•™ìŠµìš©, ê²€ì¦ìš© í–‰ ë²ˆí˜¸ë¥¼ arrayë¡œ ë¦¬í„´í•œë‹¤.

```python
for train_index, test_index in kfold.split(features):
    print(train_index)
    print(test_index)
    print("=" * 80)
```

```python
count = 0
cv_accuracy = []

for train_index, test_index in kfold.split(features):
    count += 1
    
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = targets[train_index], targets[test_index]
    
    train_targets = pd.DataFrame(y_train)
    test_targets = pd.DataFrame(y_test)
    
    #í•™ìŠµ ë° ì˜ˆì¸¡
    decision_tree_classifier.fit(X_train, y_train)
    prediction = decision_tree_classifier.predict(X_test)
    
    # ì •í™•ë„ ì¸¡ì •
    accuracy = np.round(accuracy_score(y_test, prediction), 4)
    
    cv_accuracy.append(accuracy)
    
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print(f"\n# {count} êµì°¨ ê²€ì¦ ì •í™•ë„: {accuracy}, í•™ìŠµ ë°ì´í„° í¬ê¸°: {train_size}, ê²€ì¦ ë°ì´í„° í¬ê¸°: {test_size}")
    print(f"#{count} í•™ìŠµ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: \n{train_targets.value_counts()}")
    print(f"#{count} ê²€ì¦ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: \n{test_targets.value_counts()}")
    print(f"#{count} í•™ìŠµ ì„¸íŠ¸ ì¸ë±ìŠ¤: {train_index}")
    print(f"#{count} ê²€ì¦ ì„¸íŠ¸ ì¸ë±ìŠ¤: {test_index}")
    print("=" * 100)

# í´ë“œ ë³„ ê²€ì¦ ì •í™•ë„ë¥¼ í•©í•˜ì—¬ í‰ê·  ì •í™•ë„ ê³„ì‚°
print(f"â–¶ í‰ê·  ê²€ì¦ ì •í™•ë„: {np.mean(cv_accuracy)}")
```

##### íƒ€ê²Ÿ ë°ì´í„°ì˜ ë¶„í¬ë¥¼ ë™ì¼í•˜ê²Œ êµì°¨ ê²€ì¦ ì§„í–‰

##### StratifiedFold.split(features, targets)
- featuresì™€ targets ëª¨ë‘ ì „ë‹¬í•˜ê³  í•™ìŠµìš©, ê²€ì¦ìš© í–‰ ë²ˆí˜¸ë¥¼ arrayë¡œ ë¦¬í„´í•œë‹¤.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
import numpy as np
import pandas as pd

decision_tree_classifier = DecisionTreeClassifier(min_samples_leaf=6, random_state=124)

# 5ê°œì˜ í´ë“œ ì„¸íŠ¸ë¡œ ë¶„ë¦¬
skfold = StratifiedKFold(n_splits=5)
```

```python
count = 0
cv_accuracy = []

for train_index, test_index in skfold.split(features, targets):
    count += 1
    
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = targets[train_index], targets[test_index]
    
    train_targets = pd.DataFrame(y_train)
    test_targets = pd.DataFrame(y_test)
    
    #í•™ìŠµ ë° ì˜ˆì¸¡
    decision_tree_classifier.fit(X_train, y_train)
    prediction = decision_tree_classifier.predict(X_test)
    
    # ì •í™•ë„ ì¸¡ì •
    accuracy = np.round(accuracy_score(y_test, prediction), 4)
    
    cv_accuracy.append(accuracy)
    
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print(f"\n# {count} êµì°¨ ê²€ì¦ ì •í™•ë„: {accuracy}, í•™ìŠµ ë°ì´í„° í¬ê¸°: {train_size}, ê²€ì¦ ë°ì´í„° í¬ê¸°: {test_size}")
    print(f"#{count} í•™ìŠµ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: \n{train_targets.value_counts()}")
    print(f"#{count} ê²€ì¦ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: \n{test_targets.value_counts()}")
    print(f"#{count} í•™ìŠµ ì„¸íŠ¸ ì¸ë±ìŠ¤: {train_index}")
    print(f"#{count} ê²€ì¦ ì„¸íŠ¸ ì¸ë±ìŠ¤: {test_index}")
    print("=" * 100)

# í´ë“œ ë³„ ê²€ì¦ ì •í™•ë„ë¥¼ í•©í•˜ì—¬ í‰ê·  ì •í™•ë„ ê³„ì‚°
print(f"â–¶ í‰ê·  ê²€ì¦ ì •í™•ë„: {np.mean(cv_accuracy)}")
```

##### í¸í•˜ê²Œ ìˆ˜í–‰í•  ìˆ˜ ìˆëŠ” êµì°¨ ê²€ì¦

##### cross_val_score(estimator, x, y, cv, scoring)
- estimator: classifier ì¢…ë¥˜ ëª¨ë¸ì´ë©´ ë‚´ë¶€ì ìœ¼ë¡œ stratified K-Foldë¡œ ì§„í–‰ëœë‹¤.
- x: featuers
- y: targets
- scoring: í‰ê°€ í•¨ìˆ˜, ì •í™•ë„(accuracy)ì™¸ì— ë‹¤ë¥¸ ê²ƒì€ ë‹¤ë¥¸ ì¥ì—ì„œ ë°°ìš´ë‹¤.
- cv: í´ë“œ ì„¸íŠ¸ ê°œìˆ˜

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
decision_tree_classifier = DecisionTreeClassifier(random_state=124, min_samples_leaf=6)

features, targets = iris.data, iris.target

score = cross_val_score(decision_tree_classifier, features, targets, scoring='accuracy', cv=5)
print('êµì°¨ ê²€ì¦ë³„ ì •í™•ë„: {}'.format(score))
print('í‰ê·  ì •í™•ë„: {}'.format(np.mean(score)))
```

#### GridSearchCV(estimator, param_grid, cv, refit, return_train_score)
- estimator: í•™ìŠµí•  ëª¨ë¸ ê°ì²´ ì‘ì„±
- param_grid: dictí˜•íƒœë¡œ ì „ë‹¬í•´ì•¼ í•˜ë©°, ì£¼ìš” keyê°’ì€ max_depth, min_samples_splitì´ë‹¤.
- cv: í´ë“œ ì„¸íŠ¸ ê°œìˆ˜
- refit: ìµœì ì˜ í•˜ì´í¼ íŒŒë¼ë¯¸í„°ë¡œ ì „ë‹¬í•œ ëª¨ë¸ ê°ì²´ë¥¼ ë‹¤ì‹œ í›ˆë ¨í•˜ê³ ì í•  ë•Œ Trueë¥¼ ì „ë‹¬í•œë‹¤, ë””í´íŠ¸ëŠ” True.
- return_train_score: êµì°¨ ê²€ì¦ ì ìˆ˜ë¥¼ ê°€ì ¸ì˜¬ ì§€ì— ëŒ€í•´ True ë˜ëŠ” Falseë¥¼ ì „ë‹¬í•œë‹¤.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

# ë°ì´í„°ë¥¼ ë¡œë”©í•˜ê³  í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ë¶„ë¦¬í•œë‹¤.
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=124)
decision_tree_classifier = DecisionTreeClassifier()

# max_depth: ë…¸ë“œê°€ ìƒì„±ë˜ëŠ” ìµœëŒ€ ê¹Šì´ ìˆ˜ ì œí•œ
# min_sample_split: ìµœì†Œ ìƒ˜í”Œ ê°œìˆ˜ ì œí•œ
parameters = {'max_depth': [2, 3, 4], 'min_samples_split': [6, 7]}
```

```python
grid_decision_tree_classifier = GridSearchCV(decision_tree_classifier
                                             , param_grid=parameters
                                             , cv=3
                                             , refit=True
                                             , return_train_score=True)

grid_decision_tree_classifier.fit(X_train, y_train)
```

```python
grid_decision_tree_classifier.cv_results_
```

```python
import pandas as pd

scores_df = pd.DataFrame(grid_decision_tree_classifier.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', 
           'split0_test_score', 'split1_test_score', 'split2_test_score']]
```

```python
print(f'GridSearchCV ìµœì  íŒŒë¼ë¯¸í„°: {grid_decision_tree_classifier.best_params_}')
print(f'GridSearchCV ìµœê³  ì •í™•ë„: {grid_decision_tree_classifier.best_score_}')

prediction = grid_decision_tree_classifier.predict(X_test)
print(f'í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ ì •í™•ë„: {accuracy_score(y_test, prediction)}')

# refit ëœ ê°ì²´ëŠ” best_estimator_ë¡œ ê°€ì ¸ì˜¬ ìˆ˜ ìˆìœ¼ë©°,
# ì´ë¯¸ grid_decision_tree_classifierê°ì²´ë¥¼ GridSearchCVë¡œ ì‘ì—…í•˜ì—¬ ìƒì„±í–ˆê¸° ë•Œë¬¸ì—
# ê²°ê³¼ëŠ” ë˜‘ê°™ì´ ë‚˜ì˜¨ë‹¤.
estimator = grid_decision_tree_classifier.best_estimator_
prediction = estimator.predict(X_test)
print(f'í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ ì •í™•ë„: {accuracy_score(y_test, prediction)}')
```
