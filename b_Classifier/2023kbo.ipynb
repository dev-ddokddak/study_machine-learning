```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://www.statiz.co.kr/stat_at.php'
response = requests.get(url)
```

```python
html = response.text
soup = BeautifulSoup(html, 'html.parser')
temp = soup.find_all("table")[0]
```

```python
df = pd.DataFrame(index = range(1000), columns = ["순", "이름", "연도", "WAR", "-", "타석", "타수", "득점", "안타", "2루타", "3루타", "홈런", "루타", "타점", 
                                                 "도루", "도루실패", "볼넷", "사구", "고의사구", "삼진", "병살", "희생타", "희생플라이", "타율", "출루", "장타", 
                                                 "OPS", "wOBA", "wRC+", "WAR2", "--"])
df
```

```python
l = 0
tr_elements = temp.find_all("tr")
max_index = len(tr_elements)

for j in range(max_index):
    temp2 = tr_elements[j]
    if len(temp2.find_all("td")) == 31:
        for i in range(31):
            temp3 = temp2.find_all("td")[i]
            df.iloc[l, i] = temp3.get_text()
        l += 1

df
```
